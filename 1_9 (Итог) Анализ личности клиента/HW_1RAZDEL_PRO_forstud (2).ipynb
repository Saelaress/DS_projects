# Итоговый проект.

Анализ личности клиента — это подробный анализ клиентов компании. Он помогает компании лучше понять своих клиентов и облегчает изменение продукции в соответствии со специфическими потребностями, поведением различных типов клиентов.

Анализ личности клиента помогает компании модифицировать свой продукт в соответствии с целевыми потребителями из различных сегментов потребителей. Например, вместо того, чтобы тратить деньги на маркетинг нового продукта для каждого клиента в базе данных компании, компания может проанализировать, какой сегмент клиентов с наибольшей вероятностью купит продукт, и затем продвигать продукт только для этого сегмента.

Описание столбцов:

```python
# Функция для присвоения возрастных групп
def assign_age_group(age):
    if age <= 30:
        return '0-30'
    elif age <= 40:
        return '31-40'
    elif age <= 50:
        return '41-50'
    elif age <= 60:
        return '51-60'
    else:
        return '60+'
    
df['age_group'] = df['age'].apply(assign_age_group)
```

```python
def spending_by_age(group):
    selected_data = df[df['age_group'] == group]

    categories = selected_data.columns[7:13]  # Категории товаров
    total_spending = selected_data.iloc[:, 7:13].mean() # Средние траты

    fig, ax = plt.subplots()
    plt.barh(categories, total_spending, color='skyblue')

    plt.xlabel('Сумма трат')
    plt.ylabel('Категория товара')
    plt.title(f'Средние траты для возрастной группы "{group}"')

    plt.show()

spending_by_age('0-30')
```

```python
# 3
new_df = df
new_df['dependents'] = df['Kidhome'] + df['Teenhome']
new_df['total_spending'] = df.iloc[:, 7:13].sum(axis=1)

grouped_data = new_df.groupby('dependents').total_spending.mean().reset_index(name='mean_total_spending')

fig = px.bar(grouped_data,
             x="dependents", y="mean_total_spending", barmode='group',
             height=500, title="Средние траты клиентов по количеству иждивенцев", text_auto=True)
fig.show()
```

```python
# 4
# Создаем фигуру и подграфики
fig, axes = plt.subplots(3, 2, figsize=(10, 8))

# Выбираем данные для разных возрастных групп
education_groups = df['Education'].unique()
for i, ax in enumerate(axes.flat[:-1]):
    education_group = education_groups[i]
    selected_data = df[df['Education'] == education_group]

    categories = selected_data.columns[7:13]  # Категории товаров
    total_spending = selected_data.iloc[:, 7:13].mean() # Средние траты

    # Создаем горизонтальную столбчатую диаграмму на текущем подграфике
    ax.barh(categories, total_spending, color='skyblue')

    # Устанавливаем заголовок для каждого подграфика
    ax.set_title(f'Распределение трат для группы "{education_group}"')

# Отображаем графики
plt.tight_layout()
plt.show()
```

```python
# 6. Люди, которые покупают товар на сайте чаще берут что-то со скидкой?
# H0: Люди, которые покупают товары на сайте, берут товары по скидке так же часто, как и те, кто берет товары офлайн

online_shop = df[df['NumWebPurchases'] > 0].NumDealsPurchases
offline_shop = df[df['NumWebPurchases'] == 0].NumDealsPurchases

alpha = .05

_, pvalue = stats.ttest_ind(online_shop, offline_shop)

if pvalue < alpha:
    print(f'P-value:{pvalue:.5f}. Нулевая гипотеза отвергается')
else:
    print(f'P-value:{pvalue:.5f}. Подтверждается нулевая гипотеза')
```

```python
# 7. Зависит ли жалоба клиента от его трат?
df_7 = df
df_7['total_spending'] = df.iloc[:, 7:13].sum(axis=1)

# Вычисление коэффициента корреляции Пирсона
corr = np.corrcoef(df_7['total_spending'], df_7['Complain'])[0, 1]

print("Коэффициент корреляции Пирсона:", corr)
```

```python
# 8. Люди, принявшие предложение о покупке лишь в последней компании тратят больше, 
# чем люди, принявшие сразу предложение первой компании?

# H0: Люди, принявшие предложение о покупке лишь в последней компании тратят столько же, 
# сколько люди, принявшие сразу предложение первой компании

accepted_first = df_7[df_7['AcceptedCmp1'] == 1].total_spending
accepted_last = df_7[df_7['Response'] == 1].total_spending

alpha = .05

_, pvalue = stats.ttest_ind(accepted_first, accepted_last)

if pvalue < alpha:
    print(f'P-value:{pvalue:.5f}. Нулевая гипотеза отвергается')
else:
    print(f'P-value:{pvalue:.5f}. Подтверждается нулевая гипотеза')
    
# Вычисление средних значений
mean_accepted_first = np.mean(accepted_first)
mean_accepted_last = np.mean(accepted_last)

# Сравнение средних значений
if mean_accepted_first > mean_accepted_last:
    print("Средние значения трат принявших предложение в первой компании выше, чем в последней")
elif mean_accepted_first < mean_accepted_last:
    print("Ниже")
else:
    print("Равны")
```
